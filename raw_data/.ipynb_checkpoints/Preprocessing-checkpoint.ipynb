```python
import pandas as pd
overall=pd.read_csv('EstNumRoadDeathsPer100kBothSexes.csv')
```

```python
fem=pd.read_csv('EstNumRoadDeathsPer100kFemale.csv')
male=pd.read_csv('EstNumRoadDeathsPer100kMale.csv')
```

```python
gdp_data=pd.read_csv('GDPPerCapitaData.csv')
metadata=pd.read_csv('CountryMetadata.csv')
```

```python
overall=overall[['ParentLocation','SpatialDimValueCode','Location','Period','IsLatestYear','FactValueNumeric']]
```

```python
overall=overall.rename(columns={"ParentLocation":"Region","SpatialDimValueCode":"CountryCode","Location":"Country"})
overall=overall.rename(columns={"Period":"Year","IsLatestYear":"MostRecent","FactValueNumeric":"Value"})
```

```python
overall.head(15)
```

```python
fem=fem[['ParentLocation','SpatialDimValueCode','Location','Period','IsLatestYear','FactValueNumeric']]
fem=fem.rename(columns={"ParentLocation":"Region","SpatialDimValueCode":"CountryCode","Location":"Country"})
fem=fem.rename(columns={"Period":"Year","IsLatestYear":"MostRecent","FactValueNumeric":"Value"})
```

```python
fem.head(15)
```

```python
male=male[['ParentLocation','SpatialDimValueCode','Location','Period','IsLatestYear','FactValueNumeric']]
male=male.rename(columns={"ParentLocation":"Region","SpatialDimValueCode":"CountryCode","Location":"Country"})
male=male.rename(columns={"Period":"Year","IsLatestYear":"MostRecent","FactValueNumeric":"Value"})
```

```python
male.head(15)
```

```python
updated=pd.merge(overall,fem,how="left",on=['CountryCode','Year'])
updated=pd.merge(updated,male,how="left",on=['CountryCode','Year'])
```

```python
updated.head(10)
```

```python
updated=updated[['Region_x','CountryCode','Country_x','Year','MostRecent_x','Value_x','Value_y','Value']]
updated=updated.rename(columns={'Region_x':'Region','Country_x':'Country','MostRecent_x':'MostRecent'})
updated=updated.rename(columns={'Value_x':'Value','Value_y':'ValueF','Value':'ValueM'})
```

```python
updated.head()
```

```python
PercF=(updated['ValueM']-updated['Value'])/(updated['ValueM']-updated['ValueF'])
```

```python
PercF=PercF.fillna(0.5)
```

```python
PercF
```

```python
Female=(PercF*updated['ValueF'])/(PercF*updated['ValueF']+(1-PercF)*updated['ValueM'])
Male=((1-PercF)*updated['ValueM'])/(PercF*updated['ValueF']+(1-PercF)*updated['ValueM'])
```

```python
updated['Female']=Female
updated['Male']=Male
updated=updated[['Region','CountryCode','Country','Year','MostRecent','Value','Female','Male']]
```

```python
updated.head()
```

```python
gdp=pd.read_csv('GDPPerCapitaData.csv')
gdp=gdp[['Country Code','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']]
```

```python
gdp.head()
```

```python
gdp_vals=[]
for i in range(len(updated)):
    CountryCode=updated['CountryCode'][i]
    Year=updated['Year'][i]
    val=gdp.loc[gdp['Country Code']==CountryCode][str(Year)].values[0]
    gdp_vals.append(val)
```

```python
gdp_df=pd.DataFrame(gdp_vals,columns=['GDP'])
```

```python
updated=updated[['Region','CountryCode','Country','Year','MostRecent','Value','Female','Male']]
```

```python
updated
```

```python
gdp_df
```

```python
df = pd.concat([updated.reset_index(drop=True),gdp_df.reset_index(drop=True)], axis=1)
```

```python
df.head()
```

```python
df
```

```python
metadata=metadata[['Country Code','IncomeGroup']]
metadata=metadata.rename(columns={'Country Code':'CountryCode'})
```

```python
df=pd.merge(df,metadata,how='left',on='CountryCode')
```

```python
import pandas as pd
overall=pd.read_csv("Overall.csv")
meta=pd.read_csv("CountryMetadata.csv")
overall=overall[["CountryCode","Country","Year","MostRecent","Value","Female","Male","GDP","IncomeGroup"]]
```

```python
overall.head()
```

```python
meta.head()
```

```python
list_regions=[]
for i in range(len(overall)):
    cc=overall['CountryCode'][i]
    val=meta.loc[meta['Country Code']==cc]['Region'].values[0]
    list_regions.append(val)
```

```python
list_regions
```

```python
regions_df=pd.DataFrame(list_regions,columns=['Region'])
```

```python
overall = pd.concat([overall.reset_index(drop=True),regions_df.reset_index(drop=True)], axis=1)
```
